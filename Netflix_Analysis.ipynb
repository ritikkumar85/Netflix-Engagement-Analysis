```python
df=pd.read_csv("Netflix Engagement Dataset.csv")
```

```python
print(df.isnull().sum())
```

```python
df.dropna(subset=['Customer ID', 'Churn Status'], inplace=True)
```

```python
print("Duplicates:", df.duplicated().sum())
```

```python
# Convert numeric columns
df['Subscription Length (Months)'] = df['Subscription Length (Months)'].astype(int)
df['Daily Watch Time (Hours)'] = df['Daily Watch Time (Hours)'].astype(float)
df['Customer Satisfaction Score (1-10)'] = df['Customer Satisfaction Score (1-10)'].astype(int)

# Convert categorical columns
df['Churn Status'] = df['Churn Status'].astype('category')
df['Subscription Plan'] = df['Subscription Plan'].astype('category')
df['Payment History (On-Time/Delayed)'] = df['Payment History (On-Time/Delayed)'].astype('category')
```

```python
print(df.columns)
```

```python
df.columns = df.columns.str.strip()
```

```python
if 'Churn Status' in df.columns:
    df['Churn Status'] = df['Churn Status'].astype('category')
else:
    print("Column 'Churn Status' not found in the dataset.")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.boxplot(x=df['Daily Watch Time (Hours)'])
plt.show()
```

```python
Q1 = df['Daily Watch Time (Hours)'].quantile(0.25)
Q3 = df['Daily Watch Time (Hours)'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df = df[(df['Daily Watch Time (Hours)'] >= lower_bound) & (df['Daily Watch Time (Hours)'] <= upper_bound)]
```

```python
sns.boxplot(x=df['Daily Watch Time (Hours)'])
plt.show()
```

```python
df.to_csv("cleaned_netflix_engagement.csv", index=False)
```
